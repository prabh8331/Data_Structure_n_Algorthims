# Big O Notation

**Big O Notation** is a mathematical way to compare the efficiency of different pieces of code by analyzing their performance in terms of time and space as the input size grows.

### Time Complexity
Time complexity refers to the number of operations an algorithm takes to complete as a function of the input size. It's used to measure how the runtime of an algorithm grows as the input increases.

### Space Complexity
Space complexity refers to the amount of memory an algorithm uses in relation to the input size. It helps assess how efficiently an algorithm uses memory.

### Types of Scenarios
Big O classifies algorithm performance into different cases:
- **Ω (Omega):** The best-case scenario, representing the minimum time an algorithm will take.
- **Θ (Theta):** The average-case scenario, representing the expected time under normal conditions.
- **O (Omicron):** The worst-case scenario, representing the maximum time the algorithm could take. Big O usually refers to the worst case.

## O(n) - Linear Time Complexity

**O(n)** means that the time or number of operations required to complete the algorithm increases proportionally with the size of the input, **n**.
- If the input size doubles, the time taken to complete the algorithm also roughly doubles.
- This is typical for algorithms that perform a single pass over the input data, like looping through an array.

````python
def print_items(n):
    for i in range(n):
        print(i)

print_items(10)
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of n values
n_values = np.arange(1, 10)  # n from 1 to 100

# Calculate O(1), O(log n), O(n), and O(n^2)
n_constant = [1 for n in n_values]  # O(1)
n_logarithmic = [np.log(n) for n in n_values]  # O(log n)
n_linear = n_values  # O(n)
n_squared = [n**2 for n in n_values]  # O(n^2)

# Plot all the functions
plt.plot(n_values, n_constant, label='O(1)', color='r', marker='s')
plt.plot(n_values, n_logarithmic, label='O(log n)', color='y', marker='^')
plt.plot(n_values, n_linear, label='O(n)', color='g', marker='o')
plt.plot(n_values, n_squared, label='O(n^2)', color='b', marker='x')

# Add labels and title
plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('O(1), O(log n), O(n), and O(n^2) Time Complexities')
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
````

## Big O: Dropping Non-Dominant Terms

In the following example, we have two parts: a nested loop that runs in **O(n²)** and a single loop that runs in **O(n)**. The total time complexity is:

```
O(n²) + O(n)
```

As **n** increases, the **n²** term grows much faster than **n**. When **n** becomes large, the **n** term becomes insignificant compared to **n²**. In Big O notation, we focus on the term that dominates the growth, so we can drop the smaller, non-dominant term.

Thus, the overall time complexity simplifies to **O(n²)**.


````python
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i,j)

    for k in range(n):
        print(k)
        
print_items(10)
````

## Big): O(log n)

Consider the following sorted list:

| 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 |

The most efficient way to find an element like `1` in this list is through a method called binary search. In binary search, we repeatedly divide the list into two halves and check which half contains the target value. We continue splitting the relevant half until the target is found.

For example, to find `1`:

1. Split the list in half:  
   | 1 | 2 | 3 | 4 |  
   | 5 | 6 | 7 | 8 |
   
   Since `1` is in the first half, we discard the second half.

2. Split the first half again:  
   | 1 | 2 |  
   | 3 | 4 |

   Again, `1` is in the first half, so we discard the second half.

3. Finally, split the first half:  
   | 1 |  
   | 2 |

   We've now found `1`.

In this case, we found the element in 3 steps. Since the original list has 8 elements, and \( 2^3 = 8 \), we can calculate the time complexity as \( \log_2 8 = 3 \).

Thus, for a list of size `n`, binary search operates in \( O(\log n) \) time. This means even for a very large list, the number of steps needed to find an element is significantly reduced.

---

## O(n log n) Time Complexity

The time complexity \( O(n \log n) \) is common in sorting algorithms like merge sort and quicksort. It is the most efficient time complexity for comparison-based sorting algorithms (under general circumstances).

For example, sorting an array using merge sort involves two steps:
1. **Dividing** the array into smaller subarrays (similar to binary search, which is where the \( \log n \) factor comes from).
2. **Merging** the subarrays back into a sorted array, which requires \( O(n) \) time for each level of division.

Therefore, the overall time complexity is \( O(n \log n) \).

Note that some specialized sorting algorithms (like counting sort or radix sort) can achieve better performance under certain conditions, but \( O(n \log n) \) is the best we can do for general-purpose comparison-based sorting.


## Big O: Handling Different Input Terms

In the following code, we cannot simply state that the first `for` loop is O(n) and the second loop is also O(n), then conclude that the total time complexity is O(n) + O(n) = O(2n), and subsequently reduce it to O(n) by dropping the constant.

The reason is that `a` and `b` represent different inputs. Therefore, the correct Big O notation is O(a + b) rather than O(n), as the inputs are independent of each other.

Similarly, in the second example, where there is a nested `for` loop, the time complexity would be O(a * b), because each loop runs independently based on its input size.

````python
def print_items(a,b):
    for i in range(a):
        print(i)

    for k in range(b):
        print(k)
        
print_items(10,5)
````

````python
def print_items(a,b):
    for i in range(a):
        for k in range(b):
            print(i,k)
        
print_items(2,3)
````

## Big O: Lists

Let’s start with the list `[11, 3, 23, 7]`:

| Index | 0  | 1  | 2  | 3  |
|-------|----|----|----|----|
| Values| 11 | 3  | 23 | 7  |

### 1. Append Operation

```python
my_list.append(17)
```

| Index | 0  | 1  | 2  | 3  | 4  |
|-------|----|----|----|----|----|
| Values| 11 | 3  | 23 | 7  | 17 |

Appending adds an item to the end of the list, and no reindexing is required. This is a constant time operation, O(1).

### 2. Pop Operation (Without Index)

```python
my_list.pop()
```

| Index | 0  | 1  | 2  | 3  |
|-------|----|----|----|----|
| Values| 11 | 3  | 23 | 7  |

Popping the last element also doesn’t involve reindexing and is an O(1) operation.

### 3. Pop Operation (By Index)

```python
my_list.pop(0)
```

| Index | 1  | 2  | 3  |
|-------|----|----|----|
| Values| 3  | 23 | 7  |

Removing the element at index `0` requires shifting all elements one position to the left, which involves reindexing. Therefore, this is an O(n) operation, where `n` is the number of elements.

| Index | 0  | 1  | 2  |
|-------|----|----|----|
| Values| 3  | 23 | 7  |

### 4. Insert Operation (At Beginning)

```python
my_list.insert(0, 11)
```

| Index | 0  | 0  | 1  | 2  |
|-------|----|----|----|----|
| Values| 11 | 3  | 23 | 7  |

Inserting at index `0` requires shifting all the elements one position to the right. This reindexing makes it an O(n) operation.

| Index | 0  | 1  | 2  | 3  |
|-------|----|----|----|----|
| Values| 11 | 3  | 23 | 7  |

### 5. Insert Operation (In the Middle)

```python
my_list.insert(1, "Hi")
```

| Index | 0  | 1  | 1  | 2  | 3  |
|-------|----|----|----|----|----|
| Values| 11 | "Hi" | 3 | 23 | 7  |

Inserting in the middle requires shifting all subsequent elements, which involves reindexing. This is also an O(n) operation. Note that it is not O(1/2n) because in Big O notation, we consider the worst-case scenario and ignore constants, so the complexity remains O(n).

| Index | 0  | 1  | 2  | 3  | 4  |
|-------|----|----|----|----|----|
| Values| 11 | "Hi" | 3 | 23 | 7  |

### 6. Search Operation (By Value)

If we search for a value by iterating through the list, the time complexity is O(n), since we may have to check each element until we find the target.

### 7. Access Operation (By Index)

Accessing a value by its index is an O(1) operation because lists provide direct access to elements based on their index, using memory addresses.
